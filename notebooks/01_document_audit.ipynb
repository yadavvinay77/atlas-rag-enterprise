# Document intelligence audit

This notebook demonstrates why the PDF needs extraction-quality routing before indexing.

```python
from pathlib import Path
from enterprise_rag.ingestion import ingest_pdf

PDF = Path(r"C:\\Users\\yadav\\Downloads\\gamtad_vadharava_babat_Document.pdf")
pages = ingest_pdf(PDF, max_pages=10)
len(pages)
```

```python
for page in pages:
    print({
        "page": page.page_number,
        "method": page.extraction_method,
        "quality": round(page.quality_score, 3),
        "language": page.language_hint,
        "warnings": page.warnings,
        "preview": page.text[:120].replace("\n", " "),
    })
```

## Interview narrative

- Empty pages and legacy-font gibberish are routed to OCR.
- Every page retains extraction provenance and a quality score.
- Low-quality OCR can be sent to a human review queue instead of silently entering the index.
